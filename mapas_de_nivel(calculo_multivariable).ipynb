```python
from matplotlib import cm #para manejar colores
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return np.sin(x) + 2*np.cos(y)

resolucion = 100

x = np.linspace(-4, 4, resolucion)
y = np.linspace(-4, 4, resolucion)

x, y = np.meshgrid(x, y)

z = f(x, y)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(x, y, z, cmap=cm.cool)
fig.colorbar(surf)
plt.show()
```

```python
fig2,ax2 = plt.subplots()
level_map = np.linspace(np.min(z), np.max(z), num=resolucion)

cp = ax2.contourf(x,y,z, levels=level_map, cmap=cm.cool)
fig2.colorbar(cp)
```
