```python
import matplotlib.pyplot as plt   # librería para graficar
import numpy as np                # librería para manejo de vectores y utilidades matemáticas
```

```python
# FUnciones lineales

def f(m, x, b):
    return m*x - b

#puntos = res
res = 100
m = 3
b = 2 
x = np.linspace(-10.0, 10.0, num=res)

y = f(m,x,b)

#graficar
fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```python
# Funcion polinomica
def f(x):
    return 2*x**8 + x**4 +3*x**2 + 4
    #return 1*x**3 - 6*x**2 + 11*x**1 - 6

#puntos = res
res = 100
x = np.linspace(-10.0, 10.0, num=res)
y = f(x)

#graficar
fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```python
#Funcion trigonometrica
def f(x):
    return np.cos(x)

res = 100 #son puntos que le da suavizado para que se vea mejor
x = np.linspace(-10.0, 10.0, num=res)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```python
#Funcion exponencial
def f(x):
    return np.e**x

res = 100 #son puntos que le da suavizado para que se vea mejor
x = np.linspace(-1.0, 1.0, num=res)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```python
#Funcion logaritmica
def f(x):
    return np.log2(x)

res = 100 #son puntos que le da suavizado para que se vea mejor
x = np.linspace(.01, 256.0, num=res)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```python
#Funcion logaritmica
def f(x):
    y = np.zeros(len(x))
    for idx,x in enumerate(x):
        #print(f"idx {idx}, x {x}")
        if x>=0:
            y[idx] = 1.0
    #print(y)
    return y

res = 100 #son puntos que le da suavizado para que se vea mejor
x = np.linspace(-10.0, 10.0, num=res)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
```
